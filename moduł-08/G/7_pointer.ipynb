```
!python pointer-generator/run_summarization.py \
  --mode=decode \
  --data_path=finished_files/chunked/test_011.bin \
  --vocab_path=finished_files/vocab \
  --log_root=pretrained_model_tf1.2.1 \
  --max_enc_steps=400 \
  --max_dec_steps=110 \
  --coverage=1 \
  --single_pass=1
```

```
import os
from rouge import Rouge 
```

```
generated_location = "pretrained_model_tf1.2.1/decode_test_400maxenc_4beam_35mindec_110maxdec_ckpt-238410/decoded/"
generated_files = sorted(os.listdir(generated_location))
generated = []
for fname in generated_files:
    with open(generated_location + fname, "r") as f:
        generated.append(f.read().replace("\n", " "))
```

```
reference_location = "pretrained_model_tf1.2.1/decode_test_400maxenc_4beam_35mindec_110maxdec_ckpt-238410/reference/"
reference_files = sorted(os.listdir(reference_location))
reference = []
for fname in reference_files:
    with open(reference_location + fname, "r") as f:
        reference.append(f.read().replace("\n", " "))
```
